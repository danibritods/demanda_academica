```python
from pyxpdf import Document
from pyxpdf.xpdf import TextOutput, TextControl, page_iterator
import re

def PDF_to_string(doc_path): #should I name more specifically? Extrato_text(extrato_path)
    doc = Document(doc_path)
    control = TextControl(mode = "table")
    text_out = TextOutput(doc, control)
    return "\n".join(page_iterator(text_out))

report = PDF_to_string("../Extratos_Academicos/extrato_escolar_Daniel_Brito.pdf")
```

```python
def Subject_data(raw_data):
    name = raw_data[8:100].strip()
    data = raw_data[100:].strip().split()
    if len(data) == 2:
        return {"name":name,"credit":"-","workload":data[1],"grade":data[0],"situation":"CVD"}
    else:
        return {"name":name,"credit":data[0],"workload":data[1]+data[2],"grade":data[3],"situation":data[4]}


exp = r"[A-Z]{3}\d{5}.*"
subjects = re.findall(exp,report,re.M)
taken_subjects = {subject[:8]:Subject_data(subject) for subject in subjects}
taken_subjects
```

```python
def test_Subject_data():
  assert Subject_data('PRO01121     Introdução à Probabilidade e Estatística                                                                           9,3        68') ==  {'name': 'Introdução à Probabilidade e Estatística',
  'credit': '-',
  'workload': '68',
  'grade': '9,3',
  'situation': 'CVD'}, "Incorrect format of subject lectured during COVID special period"
  
  assert Subject_data('INF01101      Introdução à Ciência da Computação                                                                2  034 (034-000-000)  9,5       APR') == {'name': 'Introdução à Ciência da Computação',
  'credit': '2',
  'workload': '034(034-000-000)',
  'grade': '9,5',
  'situation': 'APR'}, "Incorrect format of a normal period subject"

test_Subject_data()
```
